# Agent 1: Weather (Phase 2 config-driven)

This notebook stays single-agent and now reads topic/timing defaults from config.yaml via simulated_city.config.load_config().

```python
# Cell purpose: Load app configuration from config.yaml and derive weather agent runtime settings.
from __future__ import annotations

from datetime import UTC, datetime, timedelta
from simulated_city.config import load_config

cfg = load_config()
sim_cfg = cfg.simulation
if sim_cfg is None:
    raise ValueError("Missing simulation config. Add simulation.weather_agent in config.yaml.")

weather_cfg = sim_cfg.weather_agent
weather_topic = f"{cfg.mqtt.base_topic}/{weather_cfg.topic_suffix}"
publish_interval_s = weather_cfg.publish_interval_s
step_minutes = sim_cfg.timestep_minutes
start_time = sim_cfg.start_time if sim_cfg.start_time else datetime.now(UTC)

print(f"MQTT base topic: {cfg.mqtt.base_topic}")
print(f"Weather topic: {weather_topic}")
print(f"Publish interval (s): {publish_interval_s}")
print(f"Step minutes: {step_minutes}")
```

```python
# Cell purpose: Generate deterministic weather events using config-driven timing/topic values.
steps = 6
weather_cycle = ("clear", "clear", "cloudy", "rain", "cloudy", "clear")
temperature_cycle = (16.0, 16.5, 15.5, 13.0, 14.0, 15.0)

events: list[dict[str, str | float | int]] = []
for index in range(steps):
    event_time = start_time + timedelta(minutes=index * step_minutes)
    event = {
        "agent": "weather",
        "topic": weather_topic,
        "tick": index,
        "timestamp": event_time.isoformat(),
        "publish_interval_s": publish_interval_s,
        "condition": weather_cycle[index % len(weather_cycle)],
        "temperature_c": temperature_cycle[index % len(temperature_cycle)],
    }
    events.append(event)

print("Phase 2 config-driven weather events:")
for event in events:
    print(event)
```

```python
# Cell purpose: Connect to MQTT and publish each weather event with delivery verification.
import importlib
import simulated_city.mqtt as mqtt_module

# Reload module so notebook kernels pick up latest helper wrappers after code changes.
mqtt_module = importlib.reload(mqtt_module)
connect_mqtt = mqtt_module.connect_mqtt
publish_json_checked = mqtt_module.publish_json_checked

# Connect using config-driven broker settings.
client = connect_mqtt(cfg.mqtt, client_id_suffix="agent-weather")

# Publish previously generated events to the configured weather topic.
for event in events:
    publish_json_checked(client, weather_topic, event)
    print(f"Published tick={event['tick']} to {weather_topic}")

# Disconnect cleanly after publishing.
client._simcity_connector.disconnect()
```
